```python
import torch
torch.cuda.device_count()
```

```python
import numpy as np
x=np.array([1.0,2,3,4,5,6,7,8,9])
y=np.array([2,4,6,8,10,12,14,16,18])
x=torch.from_numpy(x)
x=x.detach()
x=torch.tensor(x,dtype=torch.float32)
y=torch.tensor(y,dtype=torch.float32)
print(x)
print(y)
```

```python
x=x.reshape(-1,1)
```

```python
y=y.reshape(-1,1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x,y,label='Original data',s=400,c='g')
plt.show()
```

```python
input_size=1
hidden_size=1
output_size=1
```

```python
w1=torch.rand(input_size,hidden_size,requires_grad=True)
w2=torch.rand(hidden_size,output_size,requires_grad=True)
```

```python
w1,w2
```

```python
learning_rate=0.001
```

```python
b=torch.rand(x.size())
```

```python
import time


w1=torch.rand(input_size,hidden_size,requires_grad=True)
b=torch.rand(1,requires_grad=True)

y_pred=None
for iter in range(1,100):
    y_pred=(x).mm(w1).clamp(min=0).add(b**2)
    loss=(y_pred-y).median()
    
    loss.backward()
    
    if iter%10==0:
        
        with plt.rc_context(rc={'interactive': True}):

        
    
            plt.figure(figsize=(8,8))
            plt.plot(x,y_pred.detach(),label='Original data',c='g')
            plt.scatter(x,y,label='Original data',s=400,c='r')

            plt.show()
        plt.close()
        
        
        
        


    
    with torch.no_grad():
        w1-=learning_rate*w1.grad
        w2-=learning_rate*w2.grad
        w1.grad.zero_()
        w2.grad.zero_()
```

```python
w1,b
```

```python
plt.figure(figsize=(8,8))
plt.plot(x,y_pred.detach(),label='Original data',c='g')
plt.scatter(x,y,label='Original data',s=400,c='r')

plt.show()
```

* Now using torch functionality

```python
input_size=1
hidden_size=5
output_size=1
model=torch.nn.Sequential(
            torch.nn.Linear(input_size,hidden_size),
            torch.nn.ReLU(),
            torch.nn.Linear(hidden_size,output_size))

loss_fn=torch.nn.MSELoss(reduction='sum')

optimizer=torch.optim.Adam(model.parameters(),lr=0.001)
```

```python
for iter in range(10000):
    
    model_preds=model(x)
    loss=loss_fn(model_preds,y)
    
    if iter%1000==0:
        print('Loss :',loss.item())
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
```

```python
plt.figure(figsize=(8,8))
plt.plot(x,model_preds.detach(),label='Original data',c='g')
plt.plot(x,sk_predictions,label='sklearn_preds',c='b')
plt.scatter(x,y,label='Original data',s=400,c='r')

plt.show()
```
